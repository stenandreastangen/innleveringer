<h1>Totalkostnader bilhold</h1>

```python
km = 10000 # Antall km
print("Antall kjørte km er", km)
fe = 5000 # Forsikring for elbil årlig
fb = 7500 # Forsikring for bensinbil årlig
tfa = 8.38 * 365 # Vegavgift pr dag omregnet til årlig
de = 0.2 * km * 2 # Driftsoffkostnad på elbil etter antall kjørte km
db = km * 1 # Driftsoffkostnad på bensinbil etter antall kjørte km
be = km * 0.1 # Bomkostnad på elbil etter antall kjørte km
bb = km * 0.3 # Bomkostnad på bensinbil etter antall kjørte km
print("Beregnet totalkostnad for elbil")
te = fe + tfa + de + be
print("Kostnad i NOK =", te)
print("Sammenlignet med bensinbil samme med samme kjørelengde")
tb = fb + tfa + db + bb
print("Kostnad i NOK =", tb)
```
