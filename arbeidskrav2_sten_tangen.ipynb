```python
#Oppgave 4a
data = {
    "Norge": ["Oslo", 0.634],
    "England": ["London", 8.982],
    "Frankrike": ["Paris", 2.161],
    "Italia": ["Roma", 2.873]
}
```

```python
#Oppgave 4b
land = input('I hvilket land er du født?')
if land in data:
    hovedstad = data[land][0] # henter opp hovedstad i dictionary data
    innbyggere = data[land][1] # henter opp innbyggere i dictionary data
    print('Hovedstaden i landet', land, 'heter', hovedstad, 'og har', innbyggere, 'millioner innbyggere')
else:
    print('Beklager, dette landet ligger ikke i listen, skriv på norsk og husk stor forbokstav')
```

```python
#Oppgave 4c
nytt_land = input('Navn på nytt land:') # variabel for nytt land i dictionary
nytt_land_hovedstad = input('Hovedstad i landet:') # variabel for antall innbyggere i det nye landet i dictionary
nytt_land_innbyggere = int(input('Innbyggere i landet:')) # variabel for antall innbyggere i det nye landet i dictionary 
nytt_land_innbyggere_mill = nytt_land_innbyggere / 1_000_000 # variabel innbyggere omregnet til hele millioner med 3 desimaler
nytt_land_innbyggere_mill_str = f"{nytt_land_innbyggere_mill:.3f}" # Gjør om variabel til streng med kun 3 desimaler
data[nytt_land] = [nytt_land_hovedstad, nytt_land_innbyggere_mill_str] # Legger til ny nøkkel uten å fjerne eksisterende i dictionary
print(data)
```

```python
#Oppgave 5
import math as ma
a = float(input('Skriv inn verdien på katet a:'))
b = float(input('Skriv inn verdien på katet b:'))
c_hyp = ma.sqrt(a**2 + b**2) # Regner ut lengden for hypotenusen
omkrets = 2 * ma.pi * a # Regner ut omkretsen av hel sirkel
omkrets_halv = omkrets / 2 # Regner ut omkretsen av halv sirkel
print(f'Omkretsen av hele figuren basert på a og b verdi du har tastet inn er: {b + c_hyp + omkrets_halv:.2f}')
```
